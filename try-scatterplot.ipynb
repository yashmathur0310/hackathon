```python
import pandas as pd 
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_excel('DataPanelWHR2021C2.xlsx')
```

```python
df.isnull().sum()
```

```python
df_without_null=df[(~df['Perceptions of corruption'].isnull())&(~df['Healthy life expectancy at birth'].isnull())]
```

```python
df_2020=df_without_null[df_without_null['year']==2020]
```

```python
df_2020=df_2020.sort_values('Healthy life expectancy at birth',ascending=False)[:10]
df_2020
```

```python
sns.scatterplot(y='Healthy life expectancy at birth',x='Perceptions of corruption',data=df_2020,hue='Country name')
```

```python
def get_scatter(col1,col2,year):
    df_without_null=df[(~df[col1].isnull())&(~df[col2].isnull())]
    df_2020=df_without_null[df_without_null['year']==year]
    df_2020=df_2020.sort_values('Healthy life expectancy at birth',ascending=False)[:10]
    sns.scatterplot(x=col1,y=col2,data=df_2020,hue='Country name')
```
